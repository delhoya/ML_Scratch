```python
doc_seoul = doc[doc['city'] == '서울특별시']
doc_seoul.head()
```

```python
doc_seoul.corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data = doc_seoul.corr(), annot=True, fmt = '.2f', linewidths=0.5, cmap='Reds')
```

```python
# city 컬럼에 서울 특별시 할당 
is_seoul = doc['city'] == '서울특별시'

# dong 컬럼에 등촌동 할당 
# 그 결과를 새로운 변수에 할당.
find_dong = doc['dong'] == '등촌동'
# 두가지 조건를 동시에 충족하는 데이터를 필터링하여 새로운 변수에 저장 (AND)(&)
seoul_dong = doc[is_seoul & find_dong]

# 결과를 출력.
seoul_dong.head()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data = seoul_dong.corr(), annot=True, fmt = '.2f', linewidths=0.5, cmap='Reds')
```
